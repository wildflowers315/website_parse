```python
import asyncio
from pyppeteer import launch

async def html_to_pdf(url, output_path):
    browser = await launch()
    page = await browser.newPage()
    await page.goto(url, {'waitUntil': 'networkidle2'})  # wait until page fully loads
    await page.pdf({'path': output_path, 'format': 'A4', 'printBackground': True})
    await browser.close()

url = 'https://openforis.org/collect-earth-tutorials/'
output_file = 'collect_earth_tutorials.pdf'

# Directly await the async function in Colab environment
await html_to_pdf(url, output_file)
print('PDF saved to', output_file)
```

```python
import os
from google.colab import files

# Get the current working directory, which is usually /content/ in Colab
current_directory = '/content/'

# List all files and directories in the current directory
all_items = os.listdir(current_directory)

# Filter out only files (not directories)
files_to_download = [item for item in all_items if os.path.isfile(os.path.join(current_directory, item))]

if files_to_download:
    print(f"Found {len(files_to_download)} files in {current_directory}:")
    for file_name in files_to_download:
        print(f"Downloading: {file_name}")
        files.download(file_name)
else:
    print(f"No files found in {current_directory} to download.")
```

```python
import asyncio
from pyppeteer import launch

async def save_full_page_as_pdf(url, output_path):
    browser = await launch(args=['--no-sandbox', '--disable-setuid-sandbox'])
    page = await browser.newPage()
    await page.goto(url, {'waitUntil': 'networkidle2'})

    # Click all accordion toggles
    toggles = await page.querySelectorAll('span.kt-blocks-accordion-title, button.kt-accordion-title')
    for t in toggles:
        try:
            await t.click()
            await asyncio.sleep(0.3)
        except Exception as e:
            print("Could not click:", e)

    # optional: scroll to bottom to trigger lazy-load
    await page.evaluate('window.scrollTo(0, document.body.scrollHeight);')
    await asyncio.sleep(2)

    await page.pdf({'path': output_path, 'format': 'A4', 'printBackground': True})
    await browser.close()

if __name__ == '__main__':
    url = 'https://openforis.org/collect-earth-tutorials/'
    asyncio.get_event_loop().run_until_complete(
        save_full_page_as_pdf(url, 'collect_earth_full.pdf')
    )
    print("Done")
```
